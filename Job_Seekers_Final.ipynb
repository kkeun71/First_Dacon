# 유연근로제도 및 양립근무제도 시각화

유연근로제도 및 양립근무제도 도입률 시각화 / 출처 = http://www.index.go.kr/potal/stts/idxMain/selectPoSttsIdxSearch.do?idx_cd=3025&stts_cd=302503&clas_div=&idx_sys_cd=

유연근로제도 및 양립근무제도
- 통상의 근무시간·근무일을 변경하거나 근로자와 사용자가 근로시간이나 근로장소 등을 선택·조정하여 일과 생활을 조화롭게 하고, 인력활용의 효율성을 높일 수 있는 제도
- 스마트폰, 태블릿 등 각종 스마트기기가 생활 속으로 들어오면서 굳이 회사에 출근하지 않아도 업무수행이 가능해졌고, 일하는 시간과 공간에 제약이 없는 유연근무를 통해 업무생산성을 향상시켜 기업의 경쟁력 향상
- 유연근무제는 승진, 금전적 보상과 같은 전통적인 동기부여요소 외에 업무에 대한 자기 주도성, 일과 생활의 균형(Work-Life Balance)을 높게 평가하는 젊은 인재들의 유인요소로 작용하며, 일·가정이 양립할 수 있는 근로환경을 조성

유연근로제도

- 시간선택제 : 근로자의 필요에 따라 전일제 근로자보다 짧게 일하면서 기본적인 근로조건이 보장되고, 차별이 없는 일자리
- 시차출퇴근제 : 소정근무시간을 변경하지 않으면서 근무의 시작과 종료시간을 조정하여 근무시간을 탄력적으로 운영하는 제도
- 탄력근무제 : 일이 많은 주(일)의 근로시간을 늘리는 대신, 다른 주(일)의 근로시간을 줄여 평균 근로시간을 주 40시간에 맞추는 것
- 재량근무제 : 업무의 성질에 비추어 업무 수행 방법을 근로자 재량에 위임할 필요가 있는 업무에 한하여 서면합의한 근로시간을 근무한 것으로 간주하는 유연근무제
- 재택근무제 : 근로자가 자택 등 소속 사업장이 아닌 장소 중 사용자와 합의하거나 그 승인을 받은 장소에서 소정 근로의 전부 또는 일부를 제공하는 근무형태
- 원격근무제 : 정보 기술(IT)를 활용해 원격지에서 장소에 구애받지 않고 언제 어디서나 업무를 수행하는 새로운 근무 방식

## 구인구직 데이터 준비

```python
Job_search = pd.read_csv("/content/drive/MyDrive/데이콘 대회 분석/구인구직데이터.csv", encoding="CP949")

Job_search["취업률"] = Job_search["취업건수(월)"] / Job_search["구직건수(월)"]
Job_search["구인률"] = Job_search["취업건수(월)"] / Job_search["구인인원(월)"]

Job_search["마감년월"] = pd.to_datetime(Job_search["마감년월"])

Job_search["연도"] = Job_search["마감년월"].dt.year
Job_search["월"] = Job_search["마감년월"].dt.month

Job_search_by_year_month = pd.pivot_table(Job_search, index = ['연도'], values = ['취업률'], columns = '월', aggfunc = ['mean'], fill_value=0)
Job_search_by_year_month.columns = ["1월", "2월", "3월", "4월", "5월", "6월", "7월", "8월", "9월", "10월", "11월", "12월"]

Job_search_by_year_month_2 = pd.pivot_table(Job_search, index = ['연도'], values = ['구인률'], columns = '월', aggfunc = ['mean'], fill_value=0)
Job_search_by_year_month_2.columns = ["1월", "2월", "3월", "4월", "5월", "6월", "7월", "8월", "9월", "10월", "11월", "12월"]
```

## 연도별 취업률 히트맵 2010~2021 / 2021년 12월 데이터 x

```python
fig = px.imshow(Job_search_by_year_month,
                labels=dict(x="연도별 취업률", color="취업률"),
                x=['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월',],
                y=['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021'], 
                height=700
               )
fig.update_xaxes(side="top")
fig.show()
```

차트 해석
- 연도별 취업률 히트맵 2010~2021
- 3~4월, 8~9월에 취업률이 가장 높음
- 2010년도부터 취업률이 높아지다가 2019년도부터 다시 낮아지는 양상

## 연도별 구인률 히트맵 2010~2021 / 2021년 12월 데이터 x

```python
fig = px.imshow(Job_search_by_year_month_2,
                labels=dict(x="연도별 구인률", color="구인률"),
                x=['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월',],
                y=['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021'], 
                height=700
               )
fig.update_xaxes(side="top")
fig.show()
```

차트 해석
- 연도별 구인률 히트맵 2010~2021
- 3~4월에 구인률이 가장 높음
- 전년도 대비 상승하다가 2021년도를 들어가면서 구인률 하락

```python
Job_search_by_year_mean = Job_search.groupby("연도").mean()

Job_search_by_year_mean = Job_search_by_year_mean.reset_index()
```

## 연도별 취업률 라인그래프 2010~2021 / 2021년 12월 데이터 x

```python
fig = px.line(Job_search_by_year_mean, x="연도", y=["취업률", "구인률"], title='연도별 취업률 및 구인률')
fig.show()
```

차트해석
- 연도별 취업률 및 구인률 라인 그래프 (% 단위)
- 취업률은 18년도까지 꾸준히 높아지다가 하락 / 20년도부터 다시 조금씩 높아지는 경향
- 구인률은 20년도까지 높아지다가 20년도부터 다시 하락하는 경향
